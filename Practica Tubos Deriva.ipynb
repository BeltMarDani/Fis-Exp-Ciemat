Corregimos los datos con tiempo de trigger. (POR GAUSSIANAS)

Ajustamos el comienzo de la caja a una gaussiana:

```python
import math 

import scipy.optimize 
from scipy import special
```

```python
def gauss(x, expected_t, sigma, I):
    return 1/2 * I * (1 + special.erf( (x - expected_t) / (sigma * math.sqrt(2)) ))

def gauss_fit(data, title, name, p00, N):

    bin_num=200
    # extract x_data and y_data from the histogram data "data"
    hist, bin_edges = np.histogram(data, bins=bin_num)

    n = len(hist)
    x_hist = []
    for i in range(n):
        x_hist.append((bin_edges[i+1] + bin_edges[i]) / 2)

    x_hist = np.array(x_hist)
    y_hist = hist

    
    #Gaussian least-square fitting process
    param_optimised , _ = scipy.optimize.curve_fit(gauss, x_hist[:N], y_hist[:N], p0=p00, maxfev=10000)
    
    figure_features()
    plt.figure(figsize=(10,7))
    x_hist_2 = np.linspace(x_hist[0], x_hist[N], 500)
    plt.plot(x_hist_2[:N], gauss(x_hist_2, *param_optimised)[:N],'r-', label='Gaussian fit', linewidth=2)
    plt.legend()

    plt.hist(data, bins=bin_num, histtype='step')
    plt.title(title)
    plt.ylabel('Hits')
    plt.xlabel('Tiempo de deriva (ns)')
    plt.title(title)
    plt.savefig("Images/Hits_tiempos/t_trigger_Gauss/" + name + '.png')
    plt.show()

    t_0 = param_optimised[0] - 1.3 * param_optimised[1]
    return t_0

```

```python
def gauss(x, mu, sigma, C):
    return C * np.exp( -(x - mu)**2 / (2 * sigma**2))

def gauss_fit(data, title, name, p00):

    bin_num=200
    # extract x_data and y_data from the histogram data "data"
    hist, bin_edges = np.histogram(data, bins=bin_num)

    n = len(hist)
    x_hist = []
    for i in range(n):
        x_hist.append((bin_edges[i+1] + bin_edges[i]) / 2)

    x_hist = np.array(x_hist)
    y_hist = hist

    
    #Gaussian least-square fitting process
    param_optimised , _ = scipy.optimize.curve_fit(gauss, x_hist, y_hist, p0=p00, maxfev=10000)
    print(param_optimised)
    figure_features()
    plt.figure(figsize=(10,7))
    x_hist_2 = np.linspace(x_hist[0], x_hist[len(x_hist)-1], 500)
    plt.plot(x_hist_2, gauss(x_hist_2, *param_optimised),'r-', label='Gaussian fit', linewidth=2)
    plt.legend()

    plt.hist(data, bins=bin_num, histtype='step')
    plt.title(title)
    plt.ylabel('Hits')
    plt.xlabel('Tiempo de deriva (ns)')
    plt.title(title)
    plt.savefig("Images/Hits_tiempos/t_trigger/SL_total/" + name + '.png')
    plt.show()

    
    return 
```

```python
# definimos las funciones que calculen MT1 y MT2

def MT1(t1,t2,t3):
    t0 = 330
    return t2-t0 + (t1-t0 + t3-t0) / 2

def MT2(t2,t3,t4):
    t0 = 330
    return t3-t0 + (t2-t0 + t4-t0) / 2
```
